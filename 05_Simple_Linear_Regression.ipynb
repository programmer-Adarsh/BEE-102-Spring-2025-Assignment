```python
# Importing necessary libraries
import numpy as np  # For numerical operations
import matplotlib.pyplot as plt  # For plotting graphs

# Step 1: Design data (simple linear relation with some noise)
np.random.seed(0)  # Set seed for reproducibility
X = np.linspace(0, 10, 20)  # Create 20 points evenly spaced between 0 and 10
Y = 2 * X + 1 + np.random.randn(20)  # Y = 2*X + 1 + some random noise

# Step 2: Fit a simple linear regression model manually

# Calculate means of X and Y
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# Calculate coefficients: slope (m) and intercept (c)
m = np.sum((X - X_mean) * (Y - Y_mean)) / np.sum((X - X_mean)**2)  # slope
c = Y_mean - m * X_mean  # intercept

print(f"Slope (m): {m}")
print(f"Intercept (c): {c}")

# Step 3: Make predictions
Y_pred = m * X + c  # predicted Y values using our model

# Step 4: Calculate TSS, ESS, and RSS

# Total Sum of Squares (TSS): total variation in Y
TSS = np.sum((Y - Y_mean)**2)

# Explained Sum of Squares (ESS): variation explained by model
ESS = np.sum((Y_pred - Y_mean)**2)

# Residual Sum of Squares (RSS): variation not explained (error)
RSS = np.sum((Y - Y_pred)**2)

print(f"Total Sum of Squares (TSS): {TSS}")
print(f"Explained Sum of Squares (ESS): {ESS}")
print(f"Residual Sum of Squares (RSS): {RSS}")

# Step 5: Visualize the data and the model
plt.scatter(X, Y, color='blue', label='Actual Data')  # Scatter plot of actual points
plt.plot(X, Y_pred, color='red', label='Fitted Line')  # Line plot of predictions
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()
```
